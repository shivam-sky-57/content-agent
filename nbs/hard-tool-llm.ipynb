```python
import os 
import io
import contextlib
from dotenv import load_dotenv

load_dotenv()
```

```python
GOOGLE_GEMINI_API_KEY = os.environ.get("GOOGLE_GEMINI_API_KEY")
assert GOOGLE_GEMINI_API_KEY is not None
```

```python
BRIGHTDATA_SERP_API_KEY = os.environ.get("BRIGHTDATA_SERP_API_KEY")
assert BRIGHTDATA_SERP_API_KEY is not None
```

```python
from langchain_brightdata import BrightDataSERP

serp_tool = BrightDataSERP(
    bright_data_api_key=BRIGHTDATA_SERP_API_KEY,
    search_engine="google",
    zone="serp_api1",
    country="us",
    parse_results=True
)
```

```python
serp_tool.name
```

```python
serp_tool.invoke("when was python made?")
```

```python
from langchain.chat_models import init_chat_model

model = init_chat_model("gemini-3.5-flash", model_provider="google_genai",api_key=GOOGLE_GEMINI_API_KEY)
```

```python
tools = [serp_tool]
```

```python
tool_mapping = {x.name: x for x in tools}

tool_mapping = {}
for x in tools:
    tool_mapping[x.name] = x
```

```python
llm = model.bind_tools(tools)
```

```python
ai_msg = llm.invoke("When was Python made?")
```

```python
ai_msg.tool_calls
```

```python
for tool_call_val in ai_msg.tool_calls:
    func_name = tool_call_val.get("name")
    func_args = tool_call_val.get("args")
    _tool =  tool_mapping.get(func_name)
    _id = tool_call_val.get("id")
    print(func_name, func_args, _tool)
    tool_result = _tool.invoke(func_args)
    print(tool_result)
```
