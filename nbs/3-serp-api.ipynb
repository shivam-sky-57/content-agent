```python
import os
import io
import contextlib
from dotenv import load_dotenv

load_dotenv()
```

```python
BRIGHTDATA_SERP_API_KEY = os.environ.get("BRIGHTDATA_SERP_API_KEY")
assert BRIGHTDATA_SERP_API_KEY is not None
```

```python
from langchain_brightdata import BrightDataSERP 

serp_tool = BrightDataSERP(bright_data_api_key=BRIGHTDATA_SERP_API_KEY,zone="serp_api1",search_engine="google",country="us",parse_results=True)
```

```python
buf = io.StringIO()
with contextlib.redirect_stdout(buf):
    result_data = serp_tool.invoke("Top 5 reddit communites from Python")
```

```python
result_data
```

```python
import json
data = json.loads(result_data)
```
